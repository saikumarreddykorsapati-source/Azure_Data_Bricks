```
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DoubleType
```

```
circuits_schema = StructType(fields = [
    StructField('circuitId',IntegerType(),False),
    StructField('circuitRef',StringType(),True),
    StructField('name',StringType(),True),
    StructField('location',StringType(),True),
    StructField('country',StringType(),True),
    StructField('lat',DoubleType(),True),
    StructField('lng',DoubleType(),True),
    StructField('alt',IntegerType(),True),
    StructField('url',StringType(),True)
    
])
```

```
circuits_df = spark.read.option("header",True).schema(circuits_schema).csv("dbfs:/mnt/formula1adlsgen2/raw/circuits.csv")
```

## Setp-2 : select only the required columns in 4 styles

```
circuits_selected_columns_df = circuits_df.select('circuitId',
 'circuitRef',
 'name',
 'location',
 'country',
 'lat',
 'lng',
 'alt')
```

```
display(circuits_selected_columns_df)
```

##### in the next three styles we can change the name of the column there itself using oldcolumnname.alias("newcolumnname") and also we can perform other operations as well

```
circuits_selected_columns_df_method2 = circuits_df.select(
 circuits_df.circuitId.alias("circuit_id"),
 circuits_df.circuitRef,
 circuits_df.name,
 circuits_df.location,
 circuits_df.country 
)
```

```
display(circuits_selected_columns_df_method2)
```

```
circuits_selected_columns_df_method3 = circuits_df.select(
 circuits_df["circuitId"],
 circuits_df["circuitRef"],
 circuits_df["name"],
 circuits_df["location"],
 circuits_df["country"].alias("race_country_name") 
)
```

```
display(circuits_selected_columns_df_method3)
```

```
from pyspark.sql.functions import col
```

```
circuits_selected_columns_df_method4 =  circuits_df.select(
 col("circuitId"),
 col("circuitRef").alias("circuit_ref"),
 col("name"),
 col("location"),
 col("country"),
 (col("lat") + col("lng")).alias("Sum of lat and long")
 )
```

```
display(circuits_selected_columns_df_method4)
```

##  Step-3 : Renaming the columns as required

###### withColumnRenamed('existing','new') it return a new dataframe ny renaming a existing columns

```
circuits_col_renamed_df = circuits_selected_columns_df\
.withColumnRenamed('circuitId','circuit_id')\
.withColumnRenamed('circuitRef','circuit_ref')\
.withColumnRenamed('lat','latitude')\
.withColumnRenamed('lng','longitude')\
.withColumnRenamed('alt','altitude')
```

```
display(circuits_col_renamed_df)
```

## Step-4 : add new column ingestion date to the dataframe
###### withColumn(colName, col) - it adds new column or replaces  existing column that has same name

```
from pyspark.sql.functions import current_timestamp # current_timestamp - it returns column object so we can add it to dataframe
```

```
circuits_final_df  = circuits_col_renamed_df.withColumn("ingestion_date",current_timestamp())
```

```
display(circuits_final_df)
```

##### in above example since the current_timestamp is returning the col object we're directly adding it to the data frame using withColumn() but we can't add literal values like that directly ie., withColumn('env','prod') - this failes because here it's not returning column object, inorder to add literals also to the dataframe use 
### lit(col) function, it create a column of literal value

```
from pyspark.sql.functions import lit
```

```
circuits_dummy_df  = circuits_col_renamed_df.withColumn("ingestion_date",current_timestamp()) \
.withColumn("env", lit("prod"))
```

```
display(circuits_dummy_df)
```
